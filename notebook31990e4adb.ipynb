```python
data=pd.read_csv('../input/go-to-college-dataset/data.csv')
data
```

```python
data["parent_was_in_college"].unique()
```

```python
data["type_school"]=data["type_school"].replace("Academic",0)
data["type_school"]=data["type_school"].replace("Vocational",1)
data["school_accreditation"]=data["school_accreditation"].replace("A",0)
data["school_accreditation"]=data["school_accreditation"].replace("B",1)
data["gender"]=data["gender"].replace("Male",1)
data["gender"]=data["gender"].replace("Female",0)
data["interest"]=data["interest"].replace("Not Interested",-2)
data["interest"]=data["interest"].replace("Uncertain",0)
data["interest"]=data["interest"].replace("Less Interested",-1)
data["interest"]=data["interest"].replace("Very Interested",2)
data["interest"]=data["interest"].replace("Quiet Interested",1)
data["residence"]=data["residence"].replace("Urban",0)
data["residence"]=data["residence"].replace("Rural",1)
data
```

```python
data["parent_was_in_college"]=data["parent_was_in_college"].replace("True",1)
data["parent_was_in_college"]=data["parent_was_in_college"].replace("False",0)
data["in_college"]=data["in_college"].replace(" False ",0)
data["in_college"]=data["in_college"].replace(" True ",1)

data
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
trainX ,testX,trainY ,testY=train_test_split(x ,y ,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
stander=StandardScaler()
trainX=stander.fit_transform(trainX)
testX=stander.transform(testX)
```

#                                         To compare the result of difference model 

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(trainX ,trainY)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(trainX ,trainY)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(trainX ,trainY)
```

```python
import xgboost as xgb

model = xgb.XGBClassifier()
model.fit(trainX,trainY)
preds = model.predict(testX)
print(sum(preds==testY)/len(testY))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix=confusion_matrix(testY ,model.predict(testX))

model_accurcy=(confusion_matrix[0,0]+confusion_matrix[1,1])/(confusion_matrix[0,0]+confusion_matrix[0,1]+confusion_matrix[1,0]+confusion_matrix[1,1])
print(model_accurcy)
```
